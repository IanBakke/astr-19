```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

```python
lam = 5
random_data = np.random.poisson(lam=lam, size=1000)
```

```python
plt.hist(random_data, bins=100, edgecolor='black', color='skyblue')

plt.xlabel('Number of Events')
plt.ylabel('Frequency')
plt.title(f'Histrogram of Poisson Distribution (Lambda={lam})')

plt.savefig('poisson_histogram.pdf')

plt.show()
```
