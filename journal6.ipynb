# Sine Function
Defines a function that takes the input x and outputs the sine of x using numpy.

```python
import numpy as np

def get_sin(x):
    return np.sin(x)
```

# Cosine Function
Defines a function that takes the input x and outputs the cosine of x using numpy.

```python
def get_cos(x):
    return np.cos(x)
```

# Data Tabulation
Generates 1000 evenly spaced points between 0 and 2pi for x.

```python
from tabulate import tabulate

x_values = np.linspace(0.0, 2*np.pi, 1000)

sin_values = [get_sin(x) for x in x_values]
cos_values = [get_cos(x) for x in x_values]

table_data = [(a,b,c) for a, b, c in zip(x_values, sin_values, cos_values)]

table_headers = ["x", "sin(x)", "cos(x)"]
python_table = tabulate(table_data, tablefmt="grid", headers=table_headers, floatfmt=".3f")

def table():
	print(python_table)
```
