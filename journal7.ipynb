```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 100)
```

```python
def calculate_awesomeness(data):
    return np.exp(data)
```

```python
y = calculate_awesomeness(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y)

plt.xlabel("Time [milliseconds]")
plt.ylabel("Awesomeness")
plt.title("Awesomeness Over Time")
plt.grid(True)

plt.savefig("awesomeness_plot.pdf")
plt.show()
```
