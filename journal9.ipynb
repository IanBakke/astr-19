```python
import matplotlib.pyplot as plt
import numpy as np

num_samples = 1000
num_bins = 100

random_data = np.random.rand(num_samples)
```

```python
plt.hist(random_data, bins=num_bins, color='skyblue', edgecolor='black')

plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of 1000 Uniformly Distributed Random Numbers")

plt.savefig("uniform_histogram.pdf", format='pdf')

plt.show()
```
